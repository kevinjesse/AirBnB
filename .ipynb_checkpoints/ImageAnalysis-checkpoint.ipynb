```python
import pandas as pd
left = pd.read_csv("listings-ny-summary.csv")
right = pd.read_csv("listings-ny-details.csv")
result = pd.merge(left, right, on=['id','id'])
result.shape
df = result.loc[result['room_type_x'] == 'Entire home/apt']
df = df.loc[result['price_x'] <= 210]
df = df.loc[result['price_x'] >= 90]

df
```

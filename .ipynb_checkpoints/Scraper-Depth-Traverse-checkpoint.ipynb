```python
def scraper():
    imageURLs = []
    url = 'https://www.airbnb.com/s/homes?adults=2&children=0&infants=0&guests=1&click_referer=t%3ASEE_ALL%7Csid%3A6a603b73-c735-487c-abe3-289ee89f00fc%7Cst%3AMAGAZINE_HOMES&query=new%20york%20city&room_types%5B%5D=Entire%20home%2Fapt&min_bedrooms=0&min_bathrooms=0&zoom=11&search_by_map=true&sw_lat=40.69591552825081&sw_lng=-74.14156490278883&ne_lat=40.80437003370684&ne_lng=-73.87420685981176&refinement_paths%5B%5D=%2Fhomes&allow_override%5B%5D=&s_tag=zMjDtyxP'
    driver= webdriver.Chrome("/usr/local/bin/chromedriver")
    driver.get(url)
    text = driver.page_source
    #print(text)
    soup=bs4.BeautifulSoup(text)
    urlregex = re.compile(r"https:\/\/.*\.jpg")
    images=soup.find_all('div',  {"class": "_1df8dftk"})
    imageURLs = [[] for item in images]
    #     print(imageURLs)
    pastImageURLs = []

    all_image_urls = {}
    imgNum = 1
    while (True):
        elements = driver.find_elements_by_xpath("//div[@class='_8ssblpx']")
        if not elements:
            elements = driver.find_elements_by_xpath("//div[@class='_1wcpzyga']")
        if not elements:
            elements = driver.find_elements_by_xpath("//div[@class='_1lh7ha0a']")
        if not elements:
            elements = driver.find_elements_by_xpath("//div[@class='_14csrlku']")


        if not elements:
            print("ERROR")
            break

        for ind, item in enumerate(images):
            while True:
                text = driver.page_source
                soup=bs4.BeautifulSoup(text)
                imageupdate=soup.find_all('div',  {"class": "_1df8dftk"})

                t = str(imageupdate[ind])
                m = re.search(urlregex, t)
                u = m.group(0)
                if u not in imageURLs[ind]:
                    imageURLs[ind].append(u)
                    t = 1
                    hov = ActionChains(driver).move_to_element(elements[ind]).perform()
                    time.sleep(t)
                    while (True):
                        try:

                            nexts = driver.find_elements_by_xpath("//button[@type='button' and @aria-label='Next image']")
                            #WebDriverWait(driver, 3).until(EC.presence_of_element_located( nexts[en]))
                            nexts[ind].click()
                            time.sleep(t)
                            if t > 5: 
                                print("TIME ERROR")
                                break
                        except Exception:
                            t+=1
                            continue
                        t = 0
                        break
                else:
                    break
        break
        #change pages
        continue

```

```python
import pandas as pd
left = pd.read_csv("listings-ny-summary.csv")
right = pd.read_csv("listings-ny-details.csv")
result = pd.merge(left, right, on=['id','id'])
result.shape
```

```python
df = result.loc[result['room_type_x'] == 'Entire home/apt']
df
```

```python
data = {}
driver= webdriver.Chrome("/usr/local/bin/chromedriver")
for ind, listing in df.iterrows():
    listing_photos = []
    url = listing["listing_url"]
    ident = listing["id"]
    
    print(url)
    
    driver.get(url)
#     text = driver.page_source
#     #print(text)
#     soup=bs4.BeautifulSoup(text)
    image_button = driver.find_elements_by_xpath("//button[@type='button' and @class='_1inl4zjb' and @data-veloute='hero-view-photos-button']")
    image_button[0].click()
    time.sleep(1)
    
    urllist = []
    while (True):
        #media = driver.find_elements_by_xpath("//li[@class='media-photo media-photo-block text-center media-cover text-center']")
        media = driver.find_elements_by_xpath("//img[@data-veloute='slideshow-image']")
        img = media[0].get_attribute("src");
        if img not in urllist:
            urllist.append(img)
        else:
            break
        nexts = driver.find_elements_by_xpath("//button[@type='button' and @aria-label='Next']")
        #WebDriverWait(driver, 3).until(EC.presence_of_element_located( nexts[en]))
        nexts[0].click()
        time.sleep(1)
    data[ident] = urllist
```

```python
import pickle
with open('data.pkl', 'wb') as handle:
    pickle.dump(data, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
